#### 7. Cluster the pickup time of the day into 30-minute intervals (e.g. from 02:00 to 02:30)


I create a new column, `pickup_30min_interval`, to group pickup times into 30-minute intervals.

- I calculate the rounded time once by using `.dt.floor("30min")` on `tpep_pickup_datetime` and store the result in a temporary Series called `pickup_rounded`.
- I format the rounded time (`pickup_rounded`) into `HH:MM` format using `.dt.strftime("%H:%M")`.
- To create the interval, I calculate the end time by adding 30 minutes to `pickup_rounded` using `+ pd.Timedelta(minutes=30)` and format it in the same way.
- The start and end times are concatenated with a `"-"` separator to define the interval (e.g., `02:00-02:30`).

```python
pickup_rounded = df["tpep_pickup_datetime"].dt.floor("30min")

df["pickup_30min_interval"] = (
    pickup_rounded.dt.strftime("%H:%M")
    + "-"
    + (pickup_rounded + pd.Timedelta(minutes=30)).dt.strftime("%H:%M")
)

df[["tpep_pickup_datetime", "pickup_30min_interval"]].head()
```

#### 17. Build chains of trips. Two trips are consecutive in a chain if (a) they have the same VendorID, (b) the pickup location of the second trip is also the dropoff location of the first trip, (c) the pickup time of the second trip is after the dropoff time of the first trip, and (d) the pickup time of the second trip is at most 2 minutes later than the dropoff time of the first trip. Hint: Add a column chain to the dataset. A chain can have more than two trips.




To process the dataset, I start by converting the date columns into a proper datetime format, as this allows for easier manipulation later on. The data is then sorted by `VendorID` and `tpep_pickup_datetime` to ensure that the trips are processed in a logical order.

The main objective here is to assign chain IDs to trips, grouping those that are linked by matching drop-off and pick-up locations within a time window of 120 seconds. To achieve this efficiently, I convert the relevant columns into NumPy arrays, which allow for faster operations compared to direct DataFrame manipulations.

For each trip, I determine whether it already belongs to a chain. If not, I assign a new chain ID. Using binary search (`bisect_left` and `bisect_right`), I narrow down the range of potential matching trips based on the 120-second window. Within this range, I then iterate over the trips to check if the `PULocationID` matches the `DOLocationID` of the current trip. If a match is found, I propagate the chain ID to the matching trip and stop further checks using a `break`. This ensures efficiency and also resolves cases where multiple potential matches exist by assuming the closest trip in time is the correct chain match.

Finally, I update the `chain` column in the original DataFrame with the computed chain IDs and sort the data by chain for a clear view of the grouped trips.


```python
import pandas as pd
from tqdm import tqdm
import numpy as np
from bisect import bisect_left, bisect_right



# Convert datetime columns
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])
df['tpep_dropoff_datetime'] = pd.to_datetime(df['tpep_dropoff_datetime'])

# Sort the DataFrame
df = df.sort_values(by=["VendorID", "tpep_pickup_datetime"]).reset_index(drop=True)

# Initialize the `chain` column to 0
df['chain'] = 0

# Variable to track the current chain_id
current_chain_id = 1

# Convert relevant columns to NumPy arrays
vendor_ids = df['VendorID'].values
pickup_times = df['tpep_pickup_datetime'].astype(np.int64) // 10**9  # Convert to UNIX seconds
dropoff_times = df['tpep_dropoff_datetime'].astype(np.int64) // 10**9  # Convert to UNIX seconds
pu_locations = df['PULocationID'].values
do_locations = df['DOLocationID'].values
chain_array = np.zeros(len(df), dtype=np.int32)  # NumPy array to track chain_ids

# Iterate over the rows with a time window
for i in tqdm(range(len(df)), desc="Assigning chains"):
    # Assign a new chain if the current trip does not have one
    if chain_array[i] == 0:
        chain_array[i] = current_chain_id
        current_chain_id += 1

    # Get the current VendorID and dropoff time
    current_vendor = vendor_ids[i]
    current_dropoff_time = dropoff_times[i]
    current_chain_id_local = chain_array[i]

    # Find the range of relevant rows using binary search
    start_index = bisect_left(pickup_times, current_dropoff_time, lo=i+1)
    end_index = bisect_right(pickup_times, current_dropoff_time + 120, lo=i+1)

    # Iterate only over the relevant rows
    for j in range(start_index, end_index):
        # Skip if the VendorID is different
        if vendor_ids[j] != current_vendor:
            break

        # Check if the conditions for matching are met
        if (
            pu_locations[j] == do_locations[i]
        ):
            # Assign the same chain_id
            chain_array[j] = current_chain_id_local
            break  # Exit the loop when the first match is found

# Update the DataFrame with the chain values
df['chain'] = chain_array

# Sort by chain and display results
df_sorted_by_chain = df.sort_values(by=["chain", "tpep_pickup_datetime"])
print(df_sorted_by_chain[[
    "VendorID", "tpep_pickup_datetime", "tpep_dropoff_datetime", "PULocationID", "DOLocationID", "chain"
]].head(20))
```
